##Divide photos (or competitions) into a training, test, and validation set.

```python
import random

training_frac = 0.6
val_frac = 0.2
test_frac = 1- training_frac - val_frac

frac_dict = {"training":training_frac,"validation":val_frac,"test":test_frac}

#Load data frame from pickle file
#Load pickle file
df = pd.read_pickle(os.path.join(directory,'full_2024_selcted_photos_df.pkl'))
#df_train = pd.DataFrame()
#df_val = pd.DataFrame()
#df_test = pd.DataFrame()



def find_competitor_competitions(competitor_name, all_competitions_df):
    """Finds all competitions that contain a given competitor, avoiding duplicates.

    Args:
        competitor_name: The name of the competitor to search for.
        all_competitions_df: A Pandas DataFrame containing data for all competitions.

    Returns:
        A list of unique competition names that contain the given competitor.
    """
    competitor_competitions = []
    for index, row in all_competitions_df.iterrows():
        if competitor_name in row['person_name'] and row['competition_name'] not in competitor_competitions:
            competitor_competitions.append(row['competition_name'])
    return competitor_competitions

def expand_subframe(df,competition):
  df_out = pd.DataFrame()
  rows = df[df['competition_name'] == competition].index
   #Find all rows with that competition name
  #copy all rows associated with that competition to a new (training) dataframe
  df_out = df.loc[rows]
  return df_out

def select_set(t,remaining,denominator,verbose=False):
  df_out = pd.DataFrame()
  out_frac = 0
  if verbose: print("Goal fraction = ",frac_dict[t])

  while len(remaining) > 0 and out_frac <= frac_dict[t]-0.02:
    #put a little will room here because keeping all competitors in one set moves very large blocks
    #and can result in taking much more than the required fraction
    df_tmp = pd.DataFrame()

    #Randomly select one competition
    c = random.choice(remaining)
    if verbose: print("Selected competition = ",c)
    #Find all rows with that competition name

    #copy all rows associated with that competition to a new (training) dataframe
    df_tmp = expand_subframe(df,c)
    if verbose: print(df_tmp.shape)
    df_out = pd.concat([df_out,df_tmp],ignore_index=True)
    remaining.pop(remaining.index(c))
    #For each competitior in the copied competition, find all other remaining that contain that competito
    for index, row in df_tmp.iterrows():
      #print("Row:", row)
      if verbose: print("Person Name:", row['person_name'])
      cc = find_competitor_competitions(row['person_name'],df)
      if verbose: print("Found competitior",row['person_name'],"in Competitions = ",cc)
      for c in cc:
        if c in remaining:
          if verbose: print("Removing",c,"from remaining")
          df_tmp = expand_subframe(df,c)
          df_out = pd.concat([df_out,df_tmp],ignore_index=True)
          remaining.pop(remaining.index(c))
        else:
          if verbose: print(c,"not in remaining competitions")
    #Repeat until the total number of rows in the training dataframe is training_frac of the original df
    out_frac = df_out.shape[0]/denominator
    #print(out_frac)

  return [df_out,remaining,out_frac]

diff_train = 1
diff_val = 1

while diff_train + diff_val > 0.02:
  df_rows = df.shape[0]

  out_frac = 0

  #Get a list of all unique competitions in the df
  competitions = df['competition_name'].unique().tolist()
  print(competitions)
  print(len(competitions))

  #Repeat
  print("Training ====================================================")
  df_train,competitions,f = select_set("training",competitions,df_rows)
  print("Remaining competitions \n",competitions)
  print(len(competitions))
  print(f"fraction={f}")
  diff_train = abs(f-training_frac)
  print("\n")

  #Repeat for validation set with val_frac instead of training_frac
  print("Validation ====================================================")
  df_val,competitions,f = select_set("validation",competitions,df_rows)
  print("Remaining competitions \n",competitions)
  print(len(competitions))
  print(f"fraction={f}")
  diff_val = abs(f-val_frac)
  print("\n")

  #Move remaining rows into the test df
  print("Test ===========================================================")
  df_test,competitions,f = select_set("test",competitions,df_rows)
  print("Remaining competitions \n",competitions)
  print(len(competitions))
  print(len(competitions))
  print(f"fraction={f}")
  print("\n")

#Save all three dataframes to pkl files

#df_train.head()
df_train.to_pickle(os.path.join(directory,'df_train.pkl'))
print(df_train.shape)
df_val.to_pickle(os.path.join(directory,'df_val.pkl'))
print(df_val.shape)
df_test.to_pickle(os.path.join(directory,'df_test.pkl'))
print(df_test.shape)
```
